```python
import panel as pn
from bokeh.plotting import figure
from IPython.display import display
```

```python
pn.extension()
```

```python
# prepare some data
x = [1, 2, 3]
y = [6, 7, 2]
```

```python
p = figure()
line = p.line(x=x, y=y)

print(line.data_source.data)

def hello(attr, old, new):
    # Make sure we have output indepdendent of Jupyter and whatnot
    # to see if/when this is called
    with open('tmp.log', mode='a+') as f:
        f.write(f"hello model {attr} {old} {new}\n")
        f.flush()
        print("hello model")

line.data_source.on_change('data', hello)

# Works, triggers the callback
line.data_source.data.update(x=[2, 4, 6])
```

```python
%autoreload
pane = pn.pane.Bokeh(p)
display(pane.servable())
```

```python
line.data_source.document.callbacks.unhold()
# Plot is updated, but event handler is not called
line.data_source.data.update(x=[3, 6, 9])
pn.io.push_notebook(pane)
print(line.data_source.data)
```
